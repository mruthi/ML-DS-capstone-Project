```python
descriptive=pd.DataFrame(index=["Mean","Median","Mode","Q1:25%","Q2:50%","Q3:75%","99%","Q4:100%","IQR","1.5RULE","LESSER","GREATER","MIN","MAX","kurtosis","skew","VAR","STD"],columns=Quan)
for columnName in Quan:
    descriptive[columnName]["Mean"]=dataset[columnName].mean()
    descriptive[columnName]["Median"]=dataset[columnName].median()
    descriptive[columnName]["Mode"]=dataset[columnName].mode()[0]
    descriptive[columnName]["Q1:25%"]=dataset.describe()[columnName]["25%"]
    descriptive[columnName]["Q2:50%"]=dataset.describe()[columnName]["50%"]
    descriptive[columnName]["Q3:75%"]=dataset.describe()[columnName]["75%"]
    descriptive[columnName]["99%"]=np.percentile(dataset[columnName],99)
    descriptive[columnName]["Q4:100%"]=dataset.describe()[columnName]["max"]
    descriptive[columnName]["IQR"]=descriptive[columnName]["Q3:75%"]-descriptive[columnName]["Q1:25%"]
    descriptive[columnName]["1.5RULE"]=1.5*descriptive[columnName]["IQR"]
    descriptive[columnName]["LESSER"]=descriptive[columnName]["Q1:25%"]-descriptive[columnName]["1.5RULE"]
    descriptive[columnName]["GREATER"]=descriptive[columnName]["Q3:75%"]+descriptive[columnName]["1.5RULE"]
    descriptive[columnName]["MIN"]=dataset[columnName].min()
    descriptive[columnName]["MAX"]=dataset[columnName].max()
    descriptive[columnName]["kurtosis"]=dataset[columnName].kurtosis()
    descriptive[columnName]["skew"]=dataset[columnName].skew()
    descriptive[columnName]["VAR"]=dataset[columnName].var()
    descriptive[columnName]["STD"]=dataset[columnName].std()
```

# 7)Convert the normal distribution to standard normal distribution for compactness_worst column?

```python
def stdNBgraph(dataset):
    #converting to std normal distribution
    import seaborn as sns
    mean=dataset.mean()
    std=dataset.std()
    values=[i for i in dataset]
    z_score=[((j-mean)/std)for j in values]
    sns.distplot(z_score,kde=True)
    sum(z_score)/len(z_score)
```

# 9)What is the Probability Density Function of the perimeter_worst range from 80 to 120?

```python
def get_pdf_probability(dataset,startrange,endrange):
    #importing needed libraries
    from matplotlib import pyplot
    from scipy.stats import norm
    import seaborn as sns
    #kde=kernal density fn.have to give kde=true.then only we can make curve.
    #mentioned below as curve line is blue,histogram color is green
    ax = sns.distplot(dataset,kde=True,kde_kws={'color':'blue'},color='green')
    # axis vertical line startrange color is red,endrange line color also red
    #startrange and endrange value getting from the user
    pyplot.axvline(startrange,color='Red')
    pyplot.axvline(endrange,color='Red')
    #generate a sample
    sample = dataset
    #calculating mean and median 
    sample_mean =sample.mean()
    sample_std =sample.std()
    #printing the values of mean and median
    print('Mean=%.3f, Standard Deviation=%.3f' %(sample_mean, sample_std))
    #define the distribution
    #having default fn for norm
    #its learnig how to distribute from getting mean ,median values by using norm
    dist = norm(sample_mean, sample_std)
    #sample probabilities for a range pf outcomes
    #(values = []
    #for value in range(startrange, endrange):
       # values.append(value))
    #for the above method used the below oneliner for loop method
    values = [value for value in range(startrange, endrange)]
    #values converted by using pdf fn and transformed to probabilities
    probabilities = [dist.pdf(value) for value in values]
    #creaing summation for probabilities
    prob=sum(probabilities)
    print("The area between range({},{}):{}".format(startrange,endrange,sum(probabilities)))
    return prob
```

```python
get_pdf_probability(dataset["perimeter_worst"],80,120)
```

#### RESULT:There is 0.45% Probability to get the perimeter_worst range from 80 to 120.

# 11)Detecting Multicollinearity

#### Multicollinearity refers to a situation in multiple regression analysis where two or more predictor variables are highly correlated. Multicollinearity increases the standard errors of the coefficients, making them less reliable. 

```python
#sns.pairplot(dataset)
```

##### Variance Inflation Factor (VIF): VIF measures how much the variance of a regression coefficient is inflated due to multicollinearity. A VIF value greater than 10 (some use a threshold of 5) suggests high multicollinearity.

```python
#creating calc_vif function with the input of X
#initializing an empty DataFrame of vif to store the values
#creating two columns of vif 1.variables 2.VIF
#assign the input(X)columns to the variables columns
#for loop executing,in that Input columns (shape[1]=columns,shape[0]=Rows)looping one by one to i
#VIF calculating for the x.values and i by variance_inflation_factor 
#calculated vif values stored in the list.
#function returns the vif values to the DataFrame
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(x):
    vif=pd.DataFrame()
    vif["variables"]=x.columns
    vif["VIF"]=[variance_inflation_factor(x.values,i)for i in range(x.shape[1])]
    return(vif)
```

```python
calc_vif(dataset[['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean',
       'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst',]])
```
